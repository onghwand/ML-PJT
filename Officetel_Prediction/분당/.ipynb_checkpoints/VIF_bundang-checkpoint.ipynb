```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df= pd.read_csv("bundang3years_total2.csv")
df.drop(['시군구'],axis=1,inplace=True)
df.drop(['단지명'],axis=1,inplace=True)
df.drop(['건축년도'],axis=1,inplace=True)

Y = df['거래금액'].values
Y = Y.astype('float')
X = df.drop(labels=['거래금액'],axis=1)
X = X.astype('float')

X1 = sm.add_constant(X,has_constant="add")


#표준화 파일 만들고 계수 달라지는거 확인
# minmax_scaler = MinMaxScaler()
# minmax_scaler.fit(X)
# X_scaled = minmax_scaler.transform(X)
```

```python
bun_total = sm.OLS(Y,X1)
buntotal = bun_total.fit()
buntotal.summary()
```

```python
import matplotlib.pyplot as plt
buntotal.resid.plot(label="full model")
plt.legend()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif
```
