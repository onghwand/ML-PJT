```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df= pd.read_csv("ilsan_3years_mid+large.csv")
df.drop(['시군구'],axis=1,inplace=True)
df.drop(['단지명'],axis=1,inplace=True)
df.drop(['건축년도'],axis=1,inplace=True)

Y = df['거래금액'].values
Y = Y.astype('float')
X = df.drop(labels=['거래금액'],axis=1)
X = X.astype('float')

X1 = sm.add_constant(X,has_constant="add")

#표준화
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(X)
X_scaled = minmax_scaler.transform(X)
X_scaled = sm.add_constant(X_scaled,has_constant="add")

bun_total_scaled = sm.OLS(Y,X_scaled)
buntotal_scaled = bun_total_scaled.fit()
buntotal_scaled.summary()
```
